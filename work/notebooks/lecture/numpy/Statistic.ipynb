```python
import time
```

```python
rng = np.random.default_rng(42)
shape = (1000, 1000)
a = rng.standard_normal(shape)

before = time.time()
print(a.mean())
after = time.time()
total = after - before
print(total)

print("*"*20)

before = time.time()
print(np.median(a))
after = time.time()
total = after - before
print(total)
```

```python
rng = np.random.default_rng(42)
for n in [10, 100, 1000, 10000]:
    x = rng.standard_normal((n, n))
    print(
        n, 
        f"MIN: {round(x.min(), 4)}", 
        f"MAX: {round(x.max(), 4)}"
    )
```
