## ndarrayの演算

```python
import numpy as np
```

```python
ndarray_1 = np.array([[1, 2, 3], [4, 5, 6]])
ndarray_2 = np.array([[1, 2, 3], [4, 5, 6]])
```

```python
print(ndarray_1)
print(ndarray_2)
```

```python
ndarray_1 + ndarray_2
```

```python
ndarray_1 * ndarray_2
```

```python
ndarray_3 = ([1, 2, 3])
ndarray_1 + ndarray_3
```

```python
ndarray_4 = ([1, 2])
ndarray_1 + ndarray_4
```

```python
ndarray_1 + 3
```
