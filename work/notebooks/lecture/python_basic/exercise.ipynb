```python
files = ["filename1", "filename2", "filename3"]
files
```

```python
def add_file_extension(files, extension=".png"):
    new_files = []
    for idx, file in enumerate(files):
        file_name = file + "_" + str(idx) + extension
        new_files.append(file_name)

    return new_files
```

```python
new_files = add_file_extension(files)
new_files
```

```python
def add_file_extension2(files, extension=".png"):
    new_files = []
    for idx, file in enumerate(files):
        new_files.append(f"{file}_{idx}{extension}")

    return new_files
```

```python
add_file_extension2(files)
```

```python
add_file_extension2(files, ".jpg")
```

```python
def add_file_extension3(files, extension=".png"):
    return [f"{file}_{idx}{extension}" for idx, file in enumerate(files)]
```

```python
add_file_extension3(files)
```
