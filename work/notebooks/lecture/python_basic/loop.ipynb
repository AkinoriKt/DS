# for

```python
colors = ["red", "blue", "green"]
for color in colors:
    print(color)
```

```python
for i in range(1, 11):
    print(i)
```

```python
list(range(10))
```

```python
for idx, color in enumerate(colors):
    print(idx, color)
```

# List Comprehensions

```python
colors = ["red", "blue", "green"]
colors_png = []

for color in colors:
    png_txt = color + ".png"
    colors_png.append(png_txt)

colors_png
```

```python
colors_png = [color + ".png" for color in colors]
colors_png
```

# While

```python
i = 0

while i < 5:
    print(i)
    i += 1
```
