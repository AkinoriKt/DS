```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../../../data/raw/lecture/tmdb_5000_movies.csv')
len(df)
```

```python
sns.set()
sns.jointplot(x='vote_count', y='vote_average', data=df, alpha=0.2)
```

```python
df = df[df['vote_count'] > 100]
sns.jointplot(x='vote_count', y='vote_average', data=df, alpha=0.2)
```

```python
sns.jointplot(x='vote_count', y='vote_average', 
              data=df, 
              kind='reg', 
              joint_kws={'scatter_kws': dict(alpha=0.3)})
```

```python
sns.jointplot(x='vote_count', y='vote_average', 
              data=df, 
              kind='hex')
```

```python
sns.jointplot(x='vote_count', y='vote_average', 
              data=df, 
              kind='kde')
```

```python
sns.pairplot(df[['revenue', 'budget', 'vote_count', 'vote_average']])
```

```python
df['evaluation'] = df['vote_average'].apply(lambda x: 'low' if x <= df['vote_average'].median()  else 'high')
df['evaluation']
```
