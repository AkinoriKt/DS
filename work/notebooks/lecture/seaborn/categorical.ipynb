```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('../../../data/raw/lecture/tmdb_5000_movies.csv')
```

```python
len(df)
```

```python
df = df[df['vote_average'] != 0]
len(df)
```

```python
df['vote_average'].median()
```

```python
df['evaluation'] = df['vote_average'].apply(lambda x: 'low' if x < df['vote_average'].median() else 'high')
df.head(3)
```

```python
len(df)
```

```python
sns.barplot(x='evaluation', y='revenue', data=df, estimator=np.median)
```

```python
df[df['evaluation'] == 'low']['revenue'].median()
```

```python
df.columns
```

```python
df['year'] = df['release_date'].apply(lambda x: x[:4])
df['year']
```

```python
ax = sns.countplot(x='year', data=df)
ax.tick_params(rotation=90, labelsize=6)
```
