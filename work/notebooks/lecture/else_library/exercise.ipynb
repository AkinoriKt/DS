```python
import os
import pandas as pd
from pathlib import Path
```

```python
p = Path('../../../data/raw/lecture/public-covid-data/')
```

```python
target_folder = ['rp_im', 'rp_msk']
df_list = []

for folder in p.iterdir():
    if os.path.split(folder)[-1] in target_folder:
        df_list.append(pd.DataFrame({
            'path': [sub_p.as_posix() for sub_p in folder.iterdir()],
            'filename': [os.path.split(sub_p)[-1] for sub_p in folder.iterdir()],
        }))
df = df_list[0].merge(df_list[1], on='filename', suffixes=('_im', '_msk'))
df
```

```python
p = Path('../../../data/raw/lecture/public-covid-data/')
```

```python
target_folders = {'rp_im', 'rp_msk'}
dfs = {}

for folder in p.iterdir():
    if not folder.name in target_folders:
        continue

    files = list(folder.iterdir())
    dfs[folder.name] = pd.DataFrame({
        'filename': [f.name for f in files],
        'path': [f.as_posix() for f in files],
    })

df = dfs['rp_im'].merge(
    dfs['rp_msk'],
    on='filename',
    suffixes=('_im', '_msk'),
)

df
```
