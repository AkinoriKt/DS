```python
p = Path('../../../data/raw/lecture/public-covid-data/')
p
```

```python
for sub_p in list(p.iterdir()):
    if str(sub_p).split('/')[-1] == 'rp_im':
        im_p = sub_p.iterdir()
        for p in im_p:
            im_path_list.append(p)
            path, file = os.path.split(p)
            filename_list.append(file)
    elif str(sub_p).split('/')[-1] == 'rp_msk':
        msk_p = sub_p.iterdir()
        for p in msk_p:
            msk_path_list.append(p)
```

```python
data = {
    'path_im': im_path_list,
    'filename': filename_list,
    'path_msk': msk_path_list,
}
```
