```python
import pandas as pd
import numpy as np
```

```python
df_movies = pd.read_csv('../../../data/raw/lecture/tmdb_5000_movies.csv')
df_credits = pd.read_csv('../../../data/raw/lecture/tmdb_5000_credits.csv')
```

```python
df_movies.columns
```

```python
df_credits.columns
```

```python
df = df_movies.merge(df_credits, 
                left_on='id', right_on='movie_id', 
                suffixes=('_movies', '_credits'))
df.head(3)
```

```python
df.columns
```

```python
df.describe()
```

```python
df['budget'] = df['budget'].apply(lambda x: np.nan if x == 0 else x)
df['revenue'] = df['revenue'].apply(lambda x: np.nan if x == 0 else x)
```

```python
len(df)
```

```python
df = df.dropna()
len(df)
```

```python
df.describe()
```

```python
df.columns
```

```python
df['profit'] = df['revenue'] - df['budget']
df['profit']
```

```python
df.groupby('original_language')['profit'].max()
```

```python
df.groupby('original_language')['id'].count()
```

```python
sorted_by_profit_df = df.sort_values('profit', 
                                     ascending=False).reset_index(drop=True)
sorted_by_profit_df.head()
```

```python
len(sorted_by_profit_df)
```

```python
sorted_by_profit_df.describe()
```

```python
df_movies = pd.read_csv('../../../data/raw/lecture/tmdb_5000_movies.csv')
df_credits = pd.read_csv('../../../data/raw/lecture/tmdb_5000_credits.csv')
```

```python
print(df_movies.columns)
print(df_credits.columns)
```

```python
df = df_movies.merge(df_credits, 
                     left_on='id', right_on='movie_id', 
                     suffixes=('_movies', '_credits'))
df.head(3)
```

```python
df = df[~((df['budget'] == 0) | (df['revenue'] == 0))]
len(df)
```

```python
df.describe()
```

```python
df.apply(lambda x: x['revenue'] - x['budget'], axis=1)
```

```python
df['revenue'] - df['budget']
```

```python
df['profit'] = df['revenue'] - df['budget']
df.head(3)
```

```python
len(df)
```

```python
df = df.sort_values('profit', ascending=False).reset_index(drop=True)
df.head(3)
```

```python
df.groupby('original_language').mean(numeric_only=True)
```

```python
df.to_csv(
    '../../../data/processed/lecture/tmdb_profit_sorted.csv',
    index=False
)
```
