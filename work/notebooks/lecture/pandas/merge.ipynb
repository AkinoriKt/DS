```python
data1 = {
    'key': ['k0', 'k1', 'k2'],
    'A': ['a0', 'a1', 'a2'],
    'B': ['b0', 'b1', 'b2'],
}
data2 = {
    'key': ['k0', 'k1', 'k3'],
    'C': ['c0', 'c1', 'c3'],
    'D': ['d0', 'd1', 'd3'],
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
data1 = {
    'Key': ['k0', 'k1', 'k2'],
    'ID': ['aa', 'bb', 'cc'],
    'A': ['a0', 'a1', 'a2'],
    'B': ['b0', 'b1', 'b2'],
}
data2 = {
    'Key': ['k0', 'k1', 'k3'],
    'ID': ['aa', 'bb', 'cc'],
    'C': ['c0', 'c1', 'c3'],
    'D': ['d0', 'd1', 'd3'],
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1.merge(df2, on='Key', suffixes=('_right', '_Left'))
```

```python
data1 = {
    'Key1': ['k0', 'k1', 'k2'],
    'A': ['a0', 'a1', 'a2'],
    'B': ['b0', 'b1', 'b2'],
}
data2 = {
    'Key2': ['k0', 'k1', 'k3'],
    'C': ['c0', 'c1', 'c3'],
    'D': ['d0', 'd1', 'd3'],
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1.merge(df2, left_on='Key1', right_on='Key2')
```

```python
df1.merge(df2, left_index=True, right_index=True)
```
