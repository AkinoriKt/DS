```python
import pandas as pd
```

```python
data = {
    'Date': ['Jan-1', 'Jan-1', 'Jan-1', 'Jan-2', 'Jan-2', 'Jan-2'],
    'User': ['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji', 'Robin'],
    'Method': ['Cash', 'Cash', 'Card', 'Cash', 'Card', 'Card'],
    'Price': [100, 80, 400, 250, 150, 350],
}
df = pd.DataFrame(data)
df
```

```python
df.pivot_table(values='Price', index=['Date', 'User'], columns=['Method'])
```

```python
pivod_df = df.pivot_table(values='Price', 
                          index=['Date', 'Method'], 
                          columns=['User'])
pivod_df
```

```python
pivod_df.xs('Card', level='Method')
```
