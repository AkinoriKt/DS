```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 100)
z = 1 / (1 + np.exp(-x))
```

```python
plt.plot(x, z, 
         color='r', 
         lw=3,
         ls='--',
         marker='o',
         markersize=5,
         markerfacecolor='b',
         markeredgewidth=2,
         alpha=0.4,
         label='sigmoid(x)'
)
plt.plot(x, z*2, 
         color='g', 
         lw=3,
         ls='--',
         marker='o',
         markersize=5,
         markerfacecolor='b',
         markeredgewidth=2,
         alpha=0.4,
         label='sigmoid(x*2)'
)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sigmoid(x)')
plt.legend(loc=4)
plt.xticks([1, 2, 3], rotation=20)
plt.yticks([0, 0.5, 1], rotation=20)
```
