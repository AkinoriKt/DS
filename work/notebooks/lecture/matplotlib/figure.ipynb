```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 100)
plt.subplot(1, 2, 1)
plt.plot(x, x**2)

plt.subplot(1, 2, 2)
plt.plot(x, x**3)
```

```python
fig = plt.figure()
```

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(x, x**2, label='x **2')
axes[0, 1].plot(x, x**3)
axes[1, 0].plot(x, x**4)
axes[1, 1].plot(x, x**5)

axes[0, 0].set_xlabel('xlabel')
axes[0, 0].set_ylabel('ylabel')
axes[0, 0].set_title('x ** 2')
axes[0, 0].set_xticks([-3, -1, 2])

fig.tight_layout()
axes[0, 0].legend()
axes[0, 1].axis('off')
```
