# Mutable vs Immutable

```python
list_a = [1, 2, 3]
```

```python
id(list_a)
```

```python
list_a += [4, 5, 6]
list_a
```

```python
id(list_a)
```

```python
a = 3
print(id(a))
a += 4
print(id(a))
```

```python
list_a
```

```python
def append_elem(list_param, elem=0):
    list_param.append(elem)
    return list_param
```

```python
append_elem(list_a)
print(list_a)
```

```python
list_b = [7, 8, 9]
append_elem(list_a, list_b)
```
