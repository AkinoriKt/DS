## *args, **kwargs

```python
def func(a, b):
    return a + b
```

```python
def func2(*args):
    return args[0] + args[1]
```

```python
func(1, 2)
```

```python
func2(1, 2, 3)
```

```python
files = ["filename.png", "filename.jpg", "filename.txt", "filename_2.png"]
files
```

```python
def return_only_png(files, search_extension='png'):
    png_files = []
    for file in files:
        extension = file.split('.')[-1]
        if extension == search_extension:
            png_files.append(file)

    return png_files
```

```python
return_only_png(files, 'jpg')
```

```python
def print_dict(**kwargs):
    print(kwargs)
```

```python
print_dict(a=1, b=2)
```

```python
def print_dict2(**kwargs):
    param_a = kwargs.get('a')
    print(param_a)
```

```python
print_dict2(a=1, b=2)
```

```python
def print_dict3(**kwargs):

    param_a = kwargs.get('a', 100)
    param_b = kwargs.get('b', 100)
    param_c = kwargs.get('c', 100)
    param_d = kwargs.get('d', 100)
    print(param_c)
```

```python
print_dict3(a=1000)
```

```python
def add_extension(files, extension=".png"):
    return [f"{file}_{idx}{extension}"for idx, file in enumerate(files)]
```

```python
files = ["filename1", "filename2", "filename3", "filename4"]
```

```python
png_files = add_extension(files)
png_files
```
